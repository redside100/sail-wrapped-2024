```python
import sqlite3

from pyvis.network import Network

db = sqlite3.connect("backend/wrapped.db")
rows = (
    db.cursor()
    .execute(
        "SELECT u.user_name, u.user_avatar_url, u.most_mentioned_given_name, u.most_mentioned_given_count, o.user_avatar_url FROM users u LEFT JOIN users o ON o.user_name = u.most_mentioned_given_name WHERE u.year = 2025 AND u.most_mentioned_given_count > 0"
    )
    .fetchall()
)

net = Network(
     notebook=True, height="750px", width="100%", bgcolor="#222222", font_color="white", cdn_resources = "remote"
)
for row in rows:
    username, user_avatar_url, mentioned_name, mentioned_count, mentioned_avatar_url = row
    if username not in net.get_nodes():
        net.add_node(
            username,
            label=username,
            shape="circularImage",
            image=user_avatar_url,
        )
    if mentioned_name not in net.get_nodes():
        net.add_node(
            mentioned_name,
            label=mentioned_name,
            shape="circularImage",
            image=mentioned_avatar_url,
        )
    net.add_edge(username, mentioned_name, value=mentioned_count)

net.show("network.html")
```
